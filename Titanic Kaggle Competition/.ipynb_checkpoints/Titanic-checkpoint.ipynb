```python
def cabin_clean(data):
    data['Cabin'][data['Cabin'].notnull()] = 1
    data['Cabin'][data['Cabin'].isnull()] = 0
    return data
def cabin_section(row):
    if type(row['Cabin']) == str:
        print(type(row['Cabin']))
        cabin_section = str(row['Cabin'])[0]
    else: return 'None'
    return cabin_section
#train = cabin_clean(train)
train['cabin_section'] = train.apply(cabin_section, axis=1)
test['cabin_section'] = test.apply(cabin_section, axis=1)
train.head()
```

```python
train.info()
```

```python
def embarked_clean(data):
    emb_most_freq = data['Embarked'].mode()[0]
    data['Embarked'][data['Embarked'].isnull()] = emb_most_freq
    return data
train = embarked_clean(train)
test = embarked_clean(test)
```

```python
train.info()
```

```python
def age_clean(data):
    for s in data['Sex'].unique():
        for c in data['Pclass'].unique():
            age_sc_mean = data['Age'][(data['Sex']==s) & (data['Pclass']==c)].mean()
            print(age_sc_mean, s, c)
            data['Age'][(data['Age'].isnull()) & (data['Sex']==s) & (data['Pclass']==c)] = age_sc_mean
    return data
train = age_clean(train)
test = age_clean(test)
```

```python
fare_cut_points = [-1, 10, 50, 100, 1000]
fare_labels = ['0-10', '10-50', '50-100', '100+']
train['Fare_labels'] = pd.cut(train['Fare'], fare_cut_points, labels=fare_labels)
test['Fare_labels'] = pd.cut(test['Fare'], fare_cut_points, labels=fare_labels)
```

```python
age_cut_points = [-1, 5, 12, 18, 30, 65, 100]
age_labels = ['0-5', '6-12', '13-18', '19-30', '31-65', '65+']
train['Age_labels'] = pd.cut(train['Age'], age_cut_points, labels=age_labels)
test['Age_labels'] = pd.cut(test['Age'], age_cut_points, labels=age_labels)
```

```python
dummy_df = pd.get_dummies(train[['Sex', 'Embarked', 'cabin_section', 'Fare_labels', 'Age_labels']])
dummy_df_test = pd.get_dummies(test[['Sex', 'Embarked', 'cabin_section', 'Fare_labels', 'Age_labels']])
dummy_df.head()
```

```python
train = pd.concat([train, dummy_df], axis=1)
test = pd.concat([test, dummy_df_test], axis=1)
train.head()
```

```python
train.columns
```

```python
test['cabin_section_T'] = 0
test.columns
```

```python
train_cols = ['Pclass','SibSp','Parch', 'Sex_female', 'Sex_male', 'Embarked_C',
 'Embarked_Q', 'Embarked_S', 'cabin_section_A', 'cabin_section_B','cabin_section_C', 'cabin_section_D', 'cabin_section_E',
 'cabin_section_F', 'cabin_section_G', 'cabin_section_None','cabin_section_T', 'Fare_labels_0-10', 'Fare_labels_10-50',
 'Fare_labels_50-100', 'Fare_labels_100+', 'Age_labels_0-5','Age_labels_6-12', 'Age_labels_13-18', 'Age_labels_19-30',
 'Age_labels_31-65', 'Age_labels_65+']
```

```python
train[train_cols].head()
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc = RandomForestClassifier(n_estimators=200, random_state=1, max_depth=13, min_samples_leaf=7)
rfc.fit(train[train_cols], train['Survived'])
```

```python
predictions = rfc.predict(test[train_cols])
```

```python
from sklearn.metrics import roc_auc_score
error = roc_auc_score(train["Survived"], rfc.predict(train[train_cols]))
error
```

```python
submission_df = {'PassengerId': test['PassengerId'], 'Survived': predictions}
submission = pd.DataFrame(submission_df)
submission.to_csv('titanic_predictions.csv', index=False)
```
