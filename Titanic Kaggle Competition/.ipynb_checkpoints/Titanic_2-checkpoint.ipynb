# Titanic - Machine Learning from Disaster

The Kaggle Titanic competition is an entry level type competition that provides beginner Kaggle users with an introduction of sorts to Kaggle competitions. As Titanic is my favorite movie, I found it to be quite an enjoyable experience to work through the Titanic dataset to develop a machine learning model that can determine who survived the disastrous sinking of the Titanic into the Atlantic Ocean in April of 1912. 

```python
# Import necessary python packages including pandas for cleaning and building the dataset, seaborn for plotting, 
# and a number of scikit-learn preprocessing and machine learning packages for selecting and developing a machine learning model
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import GridSearchCV, train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
import warnings
warnings.filterwarnings("ignore")
```

```python
# Create a new column that contains the first Cabin section listed in the column for each row. 
# Most of the rows that have multiple Cabin sections contain multiples of the same Cabin section 
# or a nearby section (1 letter away), so taking the first Cabin section should be effective. 
train['Cabin_Section'] = train['Cabin'].str[0]
test['Cabin_Section'] = test['Cabin'].str[0]
```

```python
# Check the first few rows to see the new Cabin_Section column addition to the dataset
train.head()
```

```python
# Inspect the value counts of the new Cabin_Section column to see that the Cabin section was properly extracted
train['Cabin_Section'].value_counts()
```

```python
# Fill the missing values in the Cabin_Section column with values of None
train['Cabin_Section'].fillna('None', inplace=True)
test['Cabin_Section'].fillna('None', inplace=True)
```

```python
# Inspect the value counts of the Cabin sections grouped by Pclass
pclass_cs_groupby = train.groupby(['Pclass', 'Cabin_Section'])['Cabin_Section'].count()
pclass_cs_groupby
```

```python
# Create a function to convert None values in Cabin_Section to the median cabin section value according to Pclass
def fill_None_cs(df):
    Pclass1_cs_med = 'C'
    Pclass2_cs_med = 'E'
    Pclass3_cs_med = 'F'
    cs_list = [Pclass1_cs_med, Pclass2_cs_med, Pclass3_cs_med]
    count=0
    for i, row in df.iterrows():
        if row['Cabin_Section'] == 'None':
            row['Cabin_Section'] = cs_list[row['Pclass']-1]
            df.iloc[i] = row
            count += 1
    return df, count
train, train_count = fill_None_cs(train)
test, test_count = fill_None_cs(test)
print(train_count, test_count)
```

```python
# Inspect the value counts of the Cabin sections grouped by Pclass again to see the change
train.groupby(['Pclass', 'Cabin_Section'])['Cabin_Section'].count()
```

```python
# Inspect the Pclass value counts
train['Pclass'].value_counts()
```

```python
# Inspect statistics of the Fare column grouped by Pclass
fare_describe = train.groupby('Pclass')['Fare'].describe()
fare_describe
```

```python
# View all rows of the train dataset where the Fare is 0
train.query('Fare==0')
```

```python
# Create a function that converts 0s in the Fare value column to the mean Fare of that row's Pclass
def fill_zeros(df):
    Pclass1_fare_avg = fare_describe.iloc[0][1]
    Pclass2_fare_avg = fare_describe.iloc[1][1]
    Pclass3_fare_avg = fare_describe.iloc[2][1]
    fare_list = [Pclass1_fare_avg, Pclass2_fare_avg, Pclass3_fare_avg]
    for i, row in df.iterrows():
        if row['Fare'] == 0:
            row['Fare'] = fare_list[row['Pclass']-1]
            df.iloc[i] = row
            print(i, row['Fare'])
    return df
train = fill_zeros(train)
test = fill_zeros(test)
```

```python
# Convert the Pclass column to a category column
train['Pclass'] = train['Pclass'].astype('category')
test['Pclass'] = test['Pclass'].astype('category')
```

```python
# Inspect the Name column values
train['Name'].head(100)
```

```python
# Create a new column that denotes whether a person has Mr, Mrs, or Miss in their name
train.loc[train['Name'].str.contains('Mr.|Ms.|Mrs.|Miss', regex=True), 'Name_title'] = 0
train.loc[~train['Name'].str.contains('Mr.|Ms.|Mrs.|Miss', regex=True), 'Name_title'] = 1

test.loc[test['Name'].str.contains('Mr.|Ms.|Mrs.|Miss', regex=True), 'Name_title'] = 0
test.loc[~test['Name'].str.contains('Mr.|Ms.|Mrs.|Miss', regex=True), 'Name_title'] = 1

train['Name_title'] = train['Name_title'].astype(int).astype('category')
test['Name_title'] = test['Name_title'].astype(int).astype('category')
```

```python
# Create new column that calculates the family size of each passenger
train['Family_size'] = train['SibSp'] + train['Parch']
test['Family_size'] = test['SibSp'] + test['Parch']
train['Family_size'].describe()
```

```python
# Create a column that calculates the length of the name of each passenger 
# since this might be an indication of passenger importance
train['Name_length'] = train['Name'].apply(len)
test['Name_length'] = test['Name'].apply(len)
```

```python
# Create dummy columns for all of the category columns
dummy_df_train = pd.get_dummies(train[['Pclass', 'Sex', 'Embarked', 'Cabin_Section', 'Age_groups', 'Fare_groups', 'Name_title']])
dummy_df_test = pd.get_dummies(test[['Pclass', 'Sex', 'Embarked', 'Cabin_Section', 'Age_groups', 'Fare_groups', 'Name_title']])
```

```python
# Concatenate the new dummy columns to the dataset
train = pd.concat([train, dummy_df_train], axis=1)
test = pd.concat([test, dummy_df_test], axis=1)
```

```python
# Inspect the first few rows of the train dataset to see the added dummy columns
train.head()
```

```python
# Get a list of train dataset columns
train.columns
```

```python
# Identify which columns should be kept for the machine learning models
cols_to_keep = ['SibSp', 'Parch', 'Pclass_1', 'Pclass_2', 'Pclass_3',
       'Sex_female', 'Sex_male', 'Embarked_C', 'Embarked_Q', 'Embarked_S',
       'Cabin_Section_A', 'Cabin_Section_B', 'Cabin_Section_C',
       'Cabin_Section_D', 'Cabin_Section_E', 'Cabin_Section_F',
       'Cabin_Section_G', 'Cabin_Section_T', 'Age_groups_(0, 4]',
       'Age_groups_(4, 12]', 'Age_groups_(12, 18]', 'Age_groups_(18, 25]',
       'Age_groups_(25, 35]', 'Age_groups_(35, 50]', 'Age_groups_(50, 65]',
       'Age_groups_(65, 120]', 'Fare_groups_(0.0, 7.8]',
       'Fare_groups_(7.8, 10.0]', 'Fare_groups_(10.0, 18.0]',
       'Fare_groups_(18.0, 25.0]', 'Fare_groups_(25.0, 30.0]',
       'Fare_groups_(30.0, 50.0]', 'Fare_groups_(50.0, 100.0]',
       'Fare_groups_(100.0, 1000.0]', 'Name_title_0', 'Name_title_1', 'Family_size', 'Name_length']
```

```python
# Identify the target column in the train dataset
train_target = train['Survived']
```

```python
# Create a new dataset containing only the columns to keep for use with the machine learning models
train_f = train[cols_to_keep]
test[['Cabin_Section_T']] = 0
test_f = test[cols_to_keep]
```

```python
# Inspect the first few rows of the newly created dataset
train_f.head()
```

```python
# Inspect the dataset datatypes and check again for null values
train_f.info()
```

```python
# Create a list of models to run preliminary machine learning predictions 
# to identify best models to move forward with GridSearchCV
models = [LogisticRegression, DecisionTreeClassifier, RandomForestClassifier, 
 GradientBoostingClassifier, SVC, KNeighborsClassifier]

# Split the train dataset into train and test datasets
train_X, test_X, train_y, test_y = train_test_split(train_f, train_target, random_state=52)

# Create a for loop to run each machine learning model on the dataset with default parameters
for model in models:
    if (model != KNeighborsClassifier):
        model = model(random_state=52)
    else: model = model()
    pipeline = make_pipeline(StandardScaler(), model)
    pipeline.fit(train_X, train_y)
    print('{} train score: '.format(str(model)) + str(pipeline.score(train_X, train_y)))
    print('{} test score: '.format(str(model)) + str(pipeline.score(test_X, test_y)) + '\n')
    #print(model.get_params())
```

```python
# Identify the models to move forward with GridSearchCV
better_models = [LogisticRegression, GradientBoostingClassifier, SVC]

# Identify the parameters for each model to loop through using GridSearchCV
parameters_LR = {'logisticregression__C': [0.01, 0.05, 0.1],
                 'logisticregression__penalty': ['l1', 'l2'],
                 'logisticregression__solver': ['liblinear', 'newton-cg', 'lbfgs', 'sag', 'saga']}
parameters_GBC = {'gradientboostingclassifier__learning_rate':[0.1, 0.15, 0.2],  
                  'gradientboostingclassifier__max_features': [6, 8, 10, 12, 15, 18],
                  'gradientboostingclassifier__min_samples_split': [4, 6, 8, 10, 12]}
parameters_SVC = {'svc__C':[1.5, 1.9, 2, 2.1, 2.2],
                  'svc__gamma': ['scale', 'auto', 0.001, 0.01, 0.1]}

parameters = [parameters_LR, parameters_GBC, parameters_SVC]
```

```python
# Create a for loop to run each model with GridSearchCV to identify the best parameters and score for each model
best_score = 0

for x in range(3):
    pipeline = make_pipeline(StandardScaler(), better_models[x](random_state=52))
    searcher = GridSearchCV(pipeline, parameters[x])
    #print(searcher.estimator.get_params().keys())
    searcher.fit(train_X, train_y)
    #print(searcher.best_score_, searcher.best_params_, '\n', 
          #searcher.best_estimator_.get_params())
    
    if searcher.best_score_ > best_score:
        best_score = searcher.best_score_
        best_estimator = searcher.best_estimator_
print('Best model: ' + '{}'.format(best_estimator) + '\n' + 'Best score: ' + '{}'.format(best_score) + '\n')
```

```python
# Use the best model to predict which ppassengers survived in the test dataset 
# to be submitted for the Titanic Kaggle competition
test_data_prediction = best_estimator.predict(test_f)
print(len(test_data_prediction))
```

```python
# Create an Excel file to be used for submittal to Kaggle
submission_df = {'PassengerId': test['PassengerId'], 'Survived': test_data_prediction}
submission = pd.DataFrame(submission_df)
submission.to_csv('titanic_predictions_2024.csv', index=False)
```
